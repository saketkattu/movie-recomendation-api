```python
import pandas as pd
import numpy as np
```

# Content Based Filtering of movie data of the TMDB5000 dataset which is available on Kaggle.

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies_df = pd.read_csv("tmdb_5000_movies.csv")
```

# The following colums represent the EDA done on the dataset

```python
credits.head()
```

```python
movies_df.head()
```

```python
credits_column_renamed=credits.rename(index=str, columns={"movie_id":"id"})
movies_df_merge = movies_df.merge(credits_column_renamed, on='id')
movies_df_merge.head()
movies_df_merge.info()
```

```python
movies_cleaned_df = movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_cleaned_df.head()
```

# Content based system using TFIDF - Document matrix from sentences present in the overview columns.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv =TfidfVectorizer(min_df=3,max_features=None,strip_accents='unicode',analyzer='word',token_pattern='\w{1,}',ngram_range=(1,3),stop_words = 'english')

movies_cleaned_df['overview']=movies_cleaned_df['overview'].fillna('') #filling Nan values
```

```python
#fitting the tfidf on the overview
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
indices=pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    idx=indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores= sorted(sig_scores, key=lambda x: x[1],reverse = True)
    sig_scores=sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
mname=input("Enter the name of your favourite movie: ")
print("We believe you will like these movies - ")
give_rec(mname)
```
